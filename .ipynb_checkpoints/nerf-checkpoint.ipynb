```python
class FrameRaysObject:

    def __init__(self, img_path, metadata, camera_angle, ray_sampling_len=64):
        self.ray_sampling_len = ray_sampling_len
        self.img = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)  # keep alpha
        if self.img.shape[2] == 4:
            alpha = self.img[:, :, 3:4] / 255.0
            rgb = self.img[:, :, :3] / 255.0
            self.img = rgb * alpha + (1 - alpha)
        focal_length = .5 * self.img.shape[1] / np.tan( camera_angle / 2)
        self.W = self.img.shape[1]
        self.H = self.img.shape[0]
        self.fx = focal_length
        self.fy = focal_length
        self.cx = self.W / 2
        self.cy = self.H / 2
        self.transform_mat = np.array(metadata["transform_matrix"])

        # creating the label imge with the true pixccels
        self.label_img = self.img.astype(np.float32).reshape(-1, 3)
        self.label_img_tensor = torch.from_numpy(self.label_img).float().to(device)
```

```python
imgs = []
transform_mats = []
fros = []
for frame_data in train_json["frames"]:
    img_path = f"./nerf_synthetic/{object_type}" + frame_data["file_path"][1:] + ".png"
    fro = FrameRaysObject(img_path, frame_data, train_json["camera_angle_x"])
    fros.append(fro)
```

```python
def get_ray_directions_img(fro):
    H, W = fro.img.shape[:2]
    i, j = np.meshgrid(np.arange(H), np.arange(W), indexing='ij')

    x = (j - fro.cx) / fro.fx
    y = -(i - fro.cy) / fro.fy
    z = np.ones_like(x)
    d_cam = np.stack([x, y, z], axis=-1)
    norms = np.linalg.norm(d_cam, axis=-1, keepdims=True)
    d_cam = d_cam / norms
    ray_directions = d_cam.reshape(-1, 3)

    return ray_directions

def convert_from_cam_to_world_coords(ray_directions, fro):
    t = np.linspace(1, 6, fro.ray_sampling_len)
    noise = np.random.uniform(0, (6-1)/fro.ray_sampling_len, size=t.shape)
    t = t + noise
    
    R = fro.transform_mat[:3, :3]
    
    ray_directions_world = ray_directions @ R.T
    return ray_directions_world, t


# def get_all_rays_img(ray_directions_world, fro, t):

#     o = torch.tensor(fro.transform_mat[:3, 3], dtype=torch.float32, device=t.device)
#     ray_directions_world = torch.tensor(ray_directions_world, dtype=torch.float32, device=t.device)

#     all_ray_pts = []

#     for ray_direction_world in ray_directions_world:
#         ray_pts = []
#         for t_val in t:
#             measured_pt = o + t_val * ray_direction_world
#             ray_pts.append(measured_pt)
#         ray_pts = torch.stack(ray_pts)
#         all_ray_pts.append(ray_pts)

#     all_ray_pts_tensor = torch.stack(all_ray_pts)
#     return all_ray_pts_tensor

def get_all_rays_img(ray_directions_world, fro, t):
    o = torch.tensor(fro.transform_mat[:3, 3], dtype=torch.float32, device=t.device)
    d = torch.tensor(ray_directions_world, dtype=torch.float32, device=t.device)
    # o: (3,), d: (N, 3), t: (64,) -> pts: (N, 64, 3)
    pts = o + t[None, :, None] * d[:, None, :]
    return pts


def positional_encoding(x, L=10):
    freqs = 2 ** torch.arange(L, dtype=torch.float32, device=x.device)
    x_enc = [x]
    for f in freqs:
        x_enc += [torch.sin(f * x), torch.cos(f * x)]
    return torch.cat(x_enc, dim=-1)  # 3 + 6*L dims

def get_inputs_labels_curr_image(fro, num_rays = 2048): # num rays is just the batch size essentially, how many rays per batch
    
    # get all ray directions
    ray_directions = get_ray_directions_img(fro)
    
    # randomly pick pixel indices
    ray_indices = np.random.choice(ray_directions.shape[0], size=num_rays, replace=False)
    
    # select only those rays
    ray_directions = ray_directions[ray_indices]
    
    # convert to world
    ray_directions_world, t = convert_from_cam_to_world_coords(ray_directions, fro)
    t = torch.tensor(t, dtype=torch.float32, device=device)
    
    # sample points along those rays
    all_ray_pts_tensor = get_all_rays_img(ray_directions_world, fro, t)
    
    # corresponding ground truth pixels
    labels = fro.label_img_tensor[ray_indices]
    
    inputs = all_ray_pts_tensor
    
    inputs = all_ray_pts_tensor.to(device)
    labels = fro.label_img_tensor[ray_indices].to(device)

    return inputs, labels, t
```

```python
img_indices = np.arange(len(fros))
```

```python
class MLP(nn.Module):
    def __init__(self, L=10):
        super().__init__()
        self.L = L
        self.mlp = nn.Sequential(
            nn.Linear(3 + 6*L, 128),
            nn.LeakyReLU(),
            nn.Linear(128, 256),
            nn.LeakyReLU(),
            nn.Linear(256, 512),
            nn.LeakyReLU(),
            nn.Linear(512, 512),
            nn.LeakyReLU(),
            nn.Linear(512, 512),
            nn.LeakyReLU(),
            nn.Linear(512, 512),
            nn.LeakyReLU(),
            nn.Linear(512, 512),
            nn.LeakyReLU(),
            nn.Linear(512, 256),
            nn.LeakyReLU(),
            nn.Linear(256, 128),
            nn.LeakyReLU(),
            nn.Linear(128, 4),
        )
        self.softplus = nn.Softplus()
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        batch_size = x.shape[0]
        x = x.reshape(-1, 3 + 6 * self.L)
        out = self.mlp(x)
        sigma = self.softplus(out[:, :1])
        rgb = self.sigmoid(out[:, 1:])
        return sigma, rgb
        
```

```python
net = MLP().to(device)
```

```python
def generate_visibility_weights(sigma, t):

    delta_t = torch.diff(t)
    delta_t = torch.cat([delta_t, delta_t[-1].reshape(-1)], dim=-1)
    ray_opacity = 1 - torch.exp(-1 * sigma[:, :, :].squeeze(-1) * delta_t).unsqueeze(-1)

    # this complicated stuff is basically
    # since the transmittence is -> 1 * (1 - opacity_0) * (1 - opacity_1) * (1 - opacity_2) ....
    # but torch.cumprod(1 - ray_opacity, dim=1) is basically (1 - opacity_0) * (1 - opacity_1) * (1 - opacity_2) ...
    # so we;re jsut prepending a 1, and removing the n-1th opacity value (1 - opacity_n-1)
    shifted_ray_opacity = torch.cat([torch.ones((ray_opacity.shape[0], 1, 1)).to(device), (1 - ray_opacity)[:, :-1, :]], dim=1)
    ray_transmittence = torch.cumprod(shifted_ray_opacity, dim=1)
    
    return ray_opacity, ray_transmittence
```

```python
optimizer = torch.optim.Adam(
    net.parameters(),
    lr=3e-4,
)
mse = nn.MSELoss()
```

```python
def calculate_loss(opacities, transmittences, labels, pred_rgb):
    weights = transmittences * opacities
    final_pred_rgb = (weights * pred_rgb).sum(dim=1)
    total_loss = mse(final_pred_rgb, labels)
    return total_loss
```

```python
def visualize_nerf_results(net, fro, device, num_rays_per_batch=2048):
    net.eval()
    H, W = fro.img.shape[:2]
    total_rays = H * W

    # get all ray directions for the full image
    ray_directions = get_ray_directions_img(fro)
    ray_directions_world, t = convert_from_cam_to_world_coords(ray_directions, fro)
    t_tensor = torch.tensor(t, dtype=torch.float32, device=device)

    all_rgb_pred = []

    with torch.no_grad():
        for i in range(0, total_rays, num_rays_per_batch):
            ray_batch = ray_directions_world[i:i+num_rays_per_batch]
            pts = get_all_rays_img(ray_batch, fro, t_tensor)  # (B, 64, 3)

            pts_flat = pts.reshape(-1, 3)
            pts_encoded = positional_encoding(pts_flat, L=10)
            pts_encoded = pts_encoded.reshape(pts.shape[0], pts.shape[1], -1)

            sigma, rgb = net(pts_encoded.float())
            sigma = sigma.reshape(-1, 64, 1)
            rgb = rgb.reshape(-1, 64, 3)

            opacities, transmittences = generate_visibility_weights(sigma, t_tensor)
            weights = transmittences * opacities
            pixel_rgb = (weights * rgb).sum(dim=1)  # (B, 3)
            all_rgb_pred.append(pixel_rgb.cpu())

    pred_img = torch.cat(all_rgb_pred, dim=0).numpy().reshape(H, W, 3)

    gt_img = fro.img[:, :, :3] if fro.img.shape[-1] == 4 else fro.img

    # compute PSNR
    mse_val = np.mean((pred_img - gt_img) ** 2)
    psnr = -10 * np.log10(mse_val) if mse_val > 0 else float('inf')

    fig, axes = plt.subplots(1, 3, figsize=(15, 5))

    axes[0].imshow(gt_img)
    axes[0].set_title("Ground Truth")
    axes[0].axis("off")

    axes[1].imshow(pred_img)
    axes[1].set_title(f"NeRF Prediction\nPSNR: {psnr:.2f} dB")
    axes[1].axis("off")

    diff = np.abs(pred_img - gt_img)
    axes[2].imshow(diff, cmap="hot")
    axes[2].set_title("Absolute Difference")
    axes[2].axis("off")

    plt.tight_layout()
    plt.show()
    net.train()
    
    return pred_img, psnr
```

```python
def compute_psnr(pred, target):
    pred = np.clip(pred, 0.0, 1.0)
    target = np.clip(target, 0.0, 1.0)
    mse_val = np.mean((pred - target) ** 2)
    if mse_val == 0:
        return float("inf")
    return -10.0 * np.log10(mse_val)

loss_history = []
psnr_history = []

num_epochs = 2000
val_fro = fros[0]  # fixed validation frame

for epoch in range(num_epochs):
    epoch_loss = 0.0
    np.random.shuffle(img_indices)

    net.train()

    for step, frame_idx in enumerate(img_indices):
        optimizer.zero_grad()

        fro = fros[frame_idx]
        inputs, labels, t = get_inputs_labels_curr_image(fro)

        inputs_flat = inputs.reshape(-1, 3)
        inputs_encoded = positional_encoding(inputs_flat, L=10)
        inputs_encoded = inputs_encoded.reshape(inputs.shape[0], inputs.shape[1], -1)

        sigma, rgb = net(inputs_encoded.float())

        sigma = sigma.reshape(-1, 64, 1)
        rgb = rgb.reshape(-1, 64, 3)

        opacities, transmittences = generate_visibility_weights(sigma, t)
        loss = calculate_loss(opacities, transmittences, labels, rgb)

        epoch_loss += loss.item()

        loss.backward()
        optimizer.step()

        if step % 10 == 0:
            print(f"[Epoch {epoch+1}] Step {step}/{len(img_indices)} | Loss: {loss.item():.6f}")

    avg_loss = epoch_loss / len(img_indices)

    print(f"\n✅ Epoch {epoch+1} completed | Avg Loss: {avg_loss:.6f}\n")

    # 💾 checkpoint every 10 epochs
    if (epoch + 1) % 10 == 0:
        # 🔥 visualization + PSNR
        pred_img, psnr = visualize_nerf_results(net, val_fro, device)
        print(f"📊 PSNR: {psnr:.2f} dB")
        checkpoint = {
            "epoch": epoch + 1,
            "model_state_dict": net.state_dict(),
            "optimizer_state_dict": optimizer.state_dict(),
            "loss": avg_loss,
        }

        torch.save(checkpoint, f"nerf_checkpoint_epoch_{epoch+1}.pth")
        print(f"💾 Saved checkpoint at epoch {epoch+1}")
```
